#### 训练集生成

```python
import numpy as np

# 设置随机种子，以便结果可重复
np.random.seed(42)

# 设置样本数
n_samples = 100

# 设置特征数
n_features = 3

# 生成随机数据 （其中可以通过乘和加来得到一个区间内的数据集）
X = np.random.randn(n_samples, n_features)

# 定义真实的权重和偏差(截距)
true_coef = np.array([3, -1,2])
true_intercept = 1

# 生成带有噪声的标签
y = np.dot(X, true_coef) + true_intercept + np.random.randn(n_samples)

# 将数据集分成训练集和测试集
train_ratio = 0.8
train_size = int(n_samples * train_ratio)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(X)
```
